### Q.6 How many patients where Readmitted ?
### Ans

```python
df['readmitted'].unique()
```

```python
df['readmitted'].value_counts()
```

##### This column It is about "days to inpatient readmission"

We'll categorize patient readmissions into two groups:

-> "Less than 30 days" will be labeled as "<30"

-> "30 days or more" will be labeled as "≥30"

##### NO -> 0
##### <30 and >30 -> 1

### Q.11 What is the highest number of medications typically prescribed to patients, and how does this correlate with the likelihood of patient readmission?
### Ans 


```python
mean= df['num_medications'].mean()
mean
```

```python
std = df['num_medications'].std()
std
```

```python
values = mean + 3*std
values                 ### upper limit
```

```python
most_common = df.groupby('readmitted')['num_medications'].agg(lambda x: x.value_counts().idxmax())
most_common
# These are the most common numbers of medicine prescribed to the admitted and non-admitted patients
```

```python
sns.displot(df, x="num_medications", hue= 'readmitted', kind="kde")
plt.title('Number of Medications VS Readmission \n\n')
plt.show()
```

## Conclusion:-
The median value of the medicine prescribed is 15.
25% of the data lies between 10-15 and 75% lies between 15-20.
Upper limit of the medication is 40 (which will be majority of highest number of medicine prescribed to patients).
And as per the Above graph we can see most READMITTED patients were prescribed 13 medicine and non-readmitted were prescribed 12. 
So we can say the patients who were prescribed more number of medicine their health condition is more critical and there chances of being readmitted is high
